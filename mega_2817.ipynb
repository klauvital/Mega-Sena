```python
from datetime import date
import numpy as np

concurso = 2817
#concurso = int(input('Digite o concurso: '))
#Converter o número para uma string
concurso_str = str(concurso)
concurso_array_digitos = np.array([int(digito) for digito in concurso_str])

# Exibir o array
print("Número do concurso:", concurso_array_digitos)
soma_concurso = np.sum(concurso_array_digitos)
print(f'soma concurso: {soma_concurso}')
#[2 8 1 7 ]


data_atual = date(2025, 1, 18) 
data_atual_str = data_atual.strftime('%d/%m/%Y')

#16/01/2025
data_str = '{}/{}/{}'.format(data_atual.day, data_atual.month, data_atual.year)

ano_string = str(data_atual.year)
# Criar um array de números inteiros a partir dos dígitos do ano
digitos_ano_atual = np.array([int(digito) for digito in str(ano_string)])
# ano [2 0 2 5]

array_hoje = np.concatenate(([data_atual.day, data_atual.month], digitos_ano_atual))
#data = np.array([1, 8, 0, 1, 2, 0, 2, 5 ])
print(array_hoje)

print('-------------------------------------')
print('Dados Concurso. Ex 2816') #
primeiro_dig_concurso = str(concurso_array_digitos[0])
print(f'primeiro_dig_concurso: {primeiro_dig_concurso}')  # 2
segundo_dig_concurso = str(concurso_array_digitos[1]) 
print(f'segundo_dig_concurso: {segundo_dig_concurso}') # 8
terceiro_dig_concurso = str(concurso_array_digitos[2])
print(f'terceiro_dig_concurso {terceiro_dig_concurso}') # 1
ultimo_dig_concurso = str(concurso_array_digitos[-1]) 
print(f'ultimo_dig_concurso {ultimo_dig_concurso}') # 6
primeira_dezena_concurso = str(concurso_array_digitos[0]) + str(concurso_array_digitos[1]) 
print(f'primeira_dezena_concurso {primeira_dezena_concurso}') # 28
segunda_dezena_concurso = str(concurso_array_digitos[2]) + str(concurso_array_digitos[3]) 
print(f'segunda_dezena_concurso {segunda_dezena_concurso}') # 16
array_segunda_dezena_concurso = ([int(digito) for digito in segunda_dezena_concurso])
print(f'array_segunda_dezena_concurso {array_segunda_dezena_concurso}') # 16
primeira_dezena_concurso_inversa = str(concurso_array_digitos[-1]) + str(concurso_array_digitos[-2]) 
print(f'primeira_dezena_concurso_inversa {primeira_dezena_concurso_inversa}')  # 61
segunda_dezena_concurso_inversa = str(concurso_array_digitos[-3]) + str(concurso_array_digitos[-4]) 
print(f'segunda_dezena_concurso_inversa {segunda_dezena_concurso_inversa}') # 82

soma_primeira_dezena_concurso = np.array([int(primeiro_dig_concurso ) + (int(segundo_dig_concurso)) ])
print(f'soma_primeira_dezena_concurso {soma_primeira_dezena_concurso}') #10
soma_segunda_dezena_concurso = np.array([int(terceiro_dig_concurso ) + (int(ultimo_dig_concurso)) ])
print(f'soma_segunda_dezena_concurso {soma_segunda_dezena_concurso}') # 7

dezena_meio_concurso = np.array([str(segundo_dig_concurso) + str(terceiro_dig_concurso)])
dezena_meio_concurso = [int(dezena_meio_concurso)] # 81
print(f'dezena_meio_concurso {dezena_meio_concurso}')
soma_dezena_meio_concurso = np.array([int(segundo_dig_concurso) + int(terceiro_dig_concurso)]) 
print(f'soma_dezena_meio_concurso {soma_dezena_meio_concurso}') # 9

print('-------------- Data--------------')


dia = np.array([int(digito) for digito in str(data_atual.day)])
print(f'dia {dia}') # 16
soma_dia = np.sum(dia)
print(f'soma_dia: {soma_dia}') # 7
mes = np.array([int(digito) for digito in str(data_atual.month)])
print(f'mes {mes}') # 1
soma_mes = np.sum(mes)
print(f'soma_mes: {soma_mes}') # 1
ano = digitos_ano_atual
print(f'digitos_ano_atual {digitos_ano_atual}') #  2 0 2 5 
soma_digitos_ano_atual = np.sum(digitos_ano_atual)
print(f'soma_digitos_ano_atual {soma_digitos_ano_atual}') # 9

primeira_dezena_ano = np.array([str(digitos_ano_atual[0] ) + str(digitos_ano_atual[1]) ])
array_primeira_dezena_ano = np.array([int(digito) for digito in primeira_dezena_ano])
print(f'primeira_dezena_ano {array_primeira_dezena_ano }') #20
segunda_dezena_ano = np.array([str(digitos_ano_atual[2] ) + str(digitos_ano_atual[3]) ])
array_segunda_dezena_ano = np.array([int(digito) for digito in segunda_dezena_ano])
print(f'segunda_dezena_ano  {array_segunda_dezena_ano }') # 25

meio_dezena_ano = np.array([str(digitos_ano_atual[1] ) + str(digitos_ano_atual[2]) ])
array_meio_dezena_ano = np.array([int(digito) for digito in meio_dezena_ano])
print(f'meio_dezena_ano {array_meio_dezena_ano}') # 02

soma_meio_dezena_ano = np.sum(array_meio_dezena_ano)
print(f'soma_meio_dezena_ano: {soma_meio_dezena_ano }') # 2

#concurso [2 8 1 6 ] data [16 01 2025] Resultado 1º 40
primeiro_numero = int(segunda_dezena_concurso) + int(primeira_dezena_ano) + int(np.array(digitos_ano_atual[0] + digitos_ano_atual[2]))
print(f'primeiro_numero: {primeiro_numero}')  #40                               


#concurso [2 8 1 6 ] data [16 01 2025] Resultado 2º 19

segundo_numero = int(soma_concurso) + int(primeiro_dig_concurso)
print(f'segundo_numero: {segundo_numero}')  #19 


#concurso [2 8 1 6 ] data [16 01 2025] Resultado 3º 17

terceiro_numero = int(soma_concurso) 
print(f'terceiro_numero: {terceiro_numero}')  #17 

#concurso [2 8 1 6 ] data [16 01 2025] Resultado 4º 04

quarto_numero = int(np.array(digitos_ano_atual[0] + digitos_ano_atual[2]))
print(f'quarto_numero: {quarto_numero}')  #04

# #concurso [2 8 1 6 ] data [16 01 2025] Resultado 5º 48 

quinto_numero = int(concurso_array_digitos[1]) * int(dia[1])
print(f'quinto_numero: {quinto_numero}')  #48


# #concurso [2 8 1 6 ] data [16 01 2025] Resultado 6º 20 

metade_1_concurso = int(concurso_array_digitos[1]) // 2
sexto_numero = int(segunda_dezena_concurso) + metade_1_concurso
print(f'sexto_numero: {sexto_numero}')  #20


numeros_do_sorteio = np.array([primeiro_numero, segundo_numero, terceiro_numero, quarto_numero, quinto_numero, sexto_numero ])
for i, numero in enumerate(numeros_do_sorteio):
    if numero > 60:
        if i != 0:
            numeros_do_sorteio[i] -= numeros_do_sorteio[i - 1]  # Subtrai o número anterior
           
        else:
            numeros_do_sorteio[i] -= numeros_do_sorteio[i + 1]  # Subtrai o número posterior
    elif numero == 0:
        if i != 0:
            numeros_do_sorteio[i] += numeros_do_sorteio[i - 1]  # Soma o número anterior
        else:
            numeros_do_sorteio[i] += numeros_do_sorteio[i + 1]  # Soma o número posterior
 

print(f'numeros do sorteio: {numeros_do_sorteio}')
# 40 – 19 – 17 – 04 – 48 – 20
numeros_do_sorteio_ordenado = np.sort(numeros_do_sorteio)
print(f'numeros do sorteio ordenado: {numeros_do_sorteio_ordenado}')
 # 4 17 19 20 40 48
print('-------------------------------')
```
